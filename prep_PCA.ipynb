```python
# loop through all images and represent them as a 4256 vector
imMat = [] # each row should be the image data for a single image
imList = [] # a list of images we're actually including
for imName in images:
    tmpName = imName
    tmpIm = cv2.imread(tmpName, cv2.IMREAD_GRAYSCALE)
    if tmpIm.shape == karolinska_shape:
        comp = tmpIm.reshape(karolinska_shape[0]*karolinska_shape[1])
        imMat.append(comp)
        imList.append(tmpName)
    else:
        print tmpIm.shape
# after this step, we have 979 images
n_samples = len(imMat)
```

```python
# center faces
faces = np.asarray(imMat)
faces_centered = faces - faces.mean(axis=0)
# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)
```

```python
def plot_gallery(title, images, n_col=5, n_row=10):
    plt.figure(figsize=(2. * n_col, 2.26 * n_row))
    plt.suptitle(title, size=16)
    for i, comp in enumerate(images):
        plt.subplot(n_row, n_col, i + 1)
        vmax = max(comp.max(), -comp.min())
        comp *= 255.0/comp.max()
        plt.imshow(comp.reshape(karolinska_shape), cmap=plt.cm.gray,
                   interpolation='nearest',
                   vmin=0,vmax=255)
                   #vmin=-vmax, vmax=vmax)
        plt.xticks(())
        plt.yticks(())
    plt.subplots_adjust(0.01, 0.05, 0.99, 0.93, 0.04, 0.)

n_components = 100
estimators = [('Eigenfaces - PCA',
     decomposition.PCA(n_components=n_components, whiten=False),
     True)]


for name, estimator, center in estimators:
    print("Extracting the top %d %s..." % (n_components, name))
    data = faces
    if center:
        data = faces_centered
    estimator.fit(data)
    if hasattr(estimator, 'cluster_centers_'):
        components_ = estimator.cluster_centers_
    else:
        components_ = estimator.components_
        
        
    #if hasattr(estimator, 'noise_variance_'):
    #    plot_gallery("Pixelwise variance",
    #                 estimator.noise_variance_.reshape(1, -1), n_col=1,
    #                 n_row=1)
    print len(components_[0])
    plot_gallery('%s - Train time %.1fs' % (name, 100),
                 components_[:50])

plt.show()
```

```python
# Now do forward selection using KNN=3 on my 100 components

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
trainDF = pd.DataFrame(train)
trainScaled = pd.DataFrame(scaler.fit_transform(trainDF), columns=trainDF.columns)
testDF = pd.DataFrame(test)
testScaled = pd.DataFrame(scaler.fit_transform(testDF), columns=testDF.columns)

```

```python
from sklearn.metrics import average_precision_score

def selectSetOfFeatures(setSize, classIndex, numNeighbors, weights, featRange):
    # forward selection code
    bestFeats = [] # a list of the best features we've seen
    candidateFeats = [i for i in range(featRange)] # 
    prevScore = -1 # the score of the composite feature set on last iteration
    trainY = np.ravel(trainScaled.iloc[:,[classIndex]].values.tolist())
    testY = testScaled.iloc[:,classIndex]
    nNeigh = numNeighbors
    #import sys

    # We'll stop iterating when we get diminishing returns on adding features
    while True:
    #while len(bestFeats) < setSize:
        bestInd   = -1 # the current best candidate feature
        bestScore = -1 # the current best score
        for feat in candidateFeats:

            neigh = KNeighborsClassifier(n_neighbors=nNeigh, weights=weights) # the classifier we're using
            if feat not in bestFeats:

                tmpFeats = list(bestFeats)
                tmpFeats.append(feat)
                X = trainScaled.iloc[:,tmpFeats]
                neigh.fit(X,trainY) 
                yPred = neigh.predict(testScaled.iloc[:,tmpFeats])
                result = np.mean([yPred[j] == testY[j] for j in range(len(yPred))])
                if result > bestScore:
                    bestScore = result
                    bestInd   = feat

        # only add the feature to list if it improves the score
        if (bestInd > -1) and (bestScore > prevScore):
            bestFeats.append(bestInd)
            prevScore = bestScore
        else:
            #precision = average_precision_score(testY,yPred)
            precision = 0
            break
    return (bestFeats, prevScore, precision)



    

def makeCompositeFromFeatures(bestFeats):
    #tmp = np.matrix([0 for i in range(17024)])
    #for feat in bestFeats:
    #    tmp = tmp+ np.matrix(components_[feat])
    #return tmp
    return np.mean([components_[feat] for feat in bestFeats],axis=0)

# so far nn=5, uniform, 100 components gives cool results

nn = 5
weight = 'uniform'
featRange = n_components
nReturn = 5
(happyFeats, happyScore, happyPrec) = selectSetOfFeatures(nReturn,happyInd,nn, weight,featRange)
happyComp = makeCompositeFromFeatures(happyFeats)
(fearFeats, fearScore, fearPrec) = selectSetOfFeatures(nReturn,fearInd,nn, weight,featRange)
fearComp = makeCompositeFromFeatures(fearFeats)
(angerFeats, angerScore, angerPrec) = selectSetOfFeatures(nReturn,angerInd,nn, weight,featRange)
angerComp = makeCompositeFromFeatures(angerFeats)
(surpriseFeats, surpriseScore, surprisePrec) = selectSetOfFeatures(nReturn,surpriseInd,nn, weight,featRange)
surpriseComp = makeCompositeFromFeatures(surpriseFeats)
(sadFeats, sadScore, sadPrec) = selectSetOfFeatures(nReturn,sadInd,nn, weight,featRange)
sadComp = makeCompositeFromFeatures(sadFeats)
(disgustFeats, disgustScore, disgustPrec) = selectSetOfFeatures(nReturn,disgustInd,nn, weight,featRange)
disgustComp = makeCompositeFromFeatures(disgustFeats)
(compFeats, compScore, compPrec) = selectSetOfFeatures(nReturn,compositeInd,nn, weight,featRange)
compComp = makeCompositeFromFeatures(compFeats)


print "happy", happyPrec
print "fear", fearPrec
print "disgust", disgustPrec
print "anger", angerPrec
print "sad", sadPrec
print "surprise", surprisePrec
            
```
